```python
# declaring an empty list
namess = []
```

```python
# appending values in the list
namess.append('Parin')
namess.append('SRK')
namess.append('Salman')
namess.append('Nisarg')
```

```python
print(namess)
```

```python
# accessing a particular element in the list (Positive Indexing)
print(namess[0])
```

```python
# accessing from reverse also (Negative Indexing)
print(namess[-1])
```

```python
# creating a list and adding items at the same time
marks = [90, 80, 80, 44, 83, 91]
marks
```

```python
# Iteration in a list
sum = 0
for m in marks:
    sum += m
    
sum
```

```python
# using len method to find the size of list
average = sum / len(marks)
average
```

```python
# merging of list
a = [1,3,5]
b = [7,9,11]
a.extend(b)
print(a)
```

```python
# manupulating list

# inserting at an index
a.insert(2, 15)
a
```

```python
# change element
a[2] = 17
a
```

```python
# pop an element at given index
a.pop(2)
a
```

```python
a.pop() # simply removes last element
a
```

```python
# remove an element using key
a.remove(9)
a
```

```python
# clear a list
a.clear()
a
```
